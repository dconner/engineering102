We are going to grab a dataset and look at some common statistics.  Maybe even draw some conclusions from our statistics. This dataset is on Heart Disease and can be found at Kaggle:

https://www.kaggle.com/datasets/neurocipher/heartdisease/data


```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("neurocipher/heartdisease")

print("Path to dataset files:", path)
```

After grabbing a dataset, lets verify an entry, looke at the dataset, and identify the columns.

```python
import os
import pandas as pd

# List files in the dataset path to find the CSV file
data_files = os.listdir(path)
csv_file = None
for f in data_files:
    if f.endswith('.csv'):
        csv_file = os.path.join(path, f)
        break

if csv_file:
    # Read the dataset into a pandas DataFrame
    df = pd.read_csv(csv_file)

    # Display the first entry
    print("\nFirst entry of the dataset:")
    print(df.head(1))

    # Print all column headings
    print("\nColumn headings:")
    print(df.columns.tolist())
else:
    print(f"No CSV file found in the dataset path: {path}")
```

Is there a number that represents that best represents a data set?

1.) We often look at at the central tendancy.  Does the Mean, Median, or Mode represent the data well?

2.) Maybe we should look at the spread of the data.
    Variance, Standard Deviation


```python
print("\n--- BP Statistics for all entries ---")
print(f"Mean BP: {df['BP'].mean():.2f}")
print(f"Median BP: {df['BP'].median():.2f}")
print(f"Mode BP: {df['BP'].mode().tolist()}")

print("\n--- BP Statistics for Sex = 0 ---")
df_sex_0 = df[df['Sex'] == 0]
print(f"Mean BP (Sex=0): {df_sex_0['BP'].mean():.2f}")
print(f"Median BP (Sex=0): {df_sex_0['BP'].median():.2f}")
print(f"Mode BP (Sex=0): {df_sex_0['BP'].mode().tolist()}")

print("\n--- BP Statistics for Sex = 1 ---")
df_sex_1 = df[df['Sex'] == 1]
print(f"Mean BP (Sex=1): {df_sex_1['BP'].mean():.2f}")
print(f"Median BP (Sex=1): {df_sex_1['BP'].median():.2f}")
print(f"Mode BP (Sex=1): {df_sex_1['BP'].mode().tolist()}")

print("\n--- BP Statistics for Heart Disease = Absence (0) ---")
df_hd_absence = df[df['Heart Disease'] == 'Absence']
if not df_hd_absence.empty:
    print(f"Mean BP (Heart Disease=Absence): {df_hd_absence['BP'].mean():.2f}")
    print(f"Median BP (Heart Disease=Absence): {df_hd_absence['BP'].median():.2f}")
    print(f"Mode BP (Heart Disease=Absence): {df_hd_absence['BP'].mode().tolist()}")
else:
    print("No entries found for Heart Disease = Absence.")

print("\n--- BP Statistics for Heart Disease = Presence (1) ---")
df_hd_presence = df[df['Heart Disease'] == 'Presence']
if not df_hd_presence.empty:
    print(f"Mean BP (Heart Disease=Presence): {df_hd_presence['BP'].mean():.2f}")
    print(f"Median BP (Heart Disease=Presence): {df_hd_presence['BP'].median():.2f}")
    print(f"Mode BP (Heart Disease=Presence): {df_hd_presence['BP'].mode().tolist()}")
else:
    print("No entries found for Heart Disease = Presence.")
```

Now the variance and standard deviation.

```python
print("\n--- BP Variance and Standard Deviation for all entries ---")
print(f"Variance BP: {df['BP'].var():.2f}")
print(f"Standard Deviation BP: {df['BP'].std():.2f}")

print("\n--- BP Variance and Standard Deviation for Sex = 0 ---")
print(f"Variance BP (Sex=0): {df_sex_0['BP'].var():.2f}")
print(f"Standard Deviation BP (Sex=0): {df_sex_0['BP'].std():.2f}")

print("\n--- BP Variance and Standard Deviation for Sex = 1 ---")
print(f"Variance BP (Sex=1): {df_sex_1['BP'].var():.2f}")
print(f"Standard Deviation BP (Sex=1): {df_sex_1['BP'].std():.2f}")

print("\n--- BP Variance and Standard Deviation for Heart Disease = Absence ---")
if not df_hd_absence.empty:
    print(f"Variance BP (Heart Disease=Absence): {df_hd_absence['BP'].var():.2f}")
    print(f"Standard Deviation BP (Heart Disease=Absence): {df_hd_absence['BP'].std():.2f}")
else:
    print("No entries found for Heart Disease = Absence.")

print("\n--- BP Variance and Standard Deviation for Heart Disease = Presence ---")
if not df_hd_presence.empty:
    print(f"Variance BP (Heart Disease=Presence): {df_hd_presence['BP'].var():.2f}")
    print(f"Standard Deviation BP (Heart Disease=Presence): {df_hd_presence['BP'].std():.2f}")
else:
    print("No entries found for Heart Disease = Presence.")
```

Correlation is about showing a relationship between two vaiables.
We will use an algorithm to see correlation.
Let's look for a relationship between BP and Cholesterol.



```python
pearson_corr = df['BP'].corr(df['Cholesterol'], method='pearson')
spearman_corr = df['BP'].corr(df['Cholesterol'], method='spearman')

print(f"Pearson correlation between BP and Cholesterol: {pearson_corr:.2f}")
print(f"Spearman correlation between BP and Cholesterol: {spearman_corr:.2f}")
```

These values suggest a very weak positive correlation between BP and Cholesterol. This means that as one variable tends to increase, the other tends to increase slightly, but the relationship is not strong. Both Pearson and Spearman correlations are close, indicating a similar weak linear and monotonic relationship.

We are going to move on to looking at Max Heartrate.  Let's look at the boxplot.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(x='Heart Disease', y='Max HR', data=df)
plt.title('Box Plot of Max Heart Rate by Heart Disease Presence')
plt.xlabel('Heart Disease')
plt.ylabel('Max Heart Rate')
plt.grid(True, alpha=0.5)
plt.show()
```

Notice the slight shift.  Is that significant?

We will look at a t-test.

T-test is a statistical tool that compares the means (averages) of one or two groups to determine if the difference between them is statistically significant or likely due to random chance

```python
from scipy import stats

# Separate Max HR for each group
max_hr_absence = df[df['Heart Disease'] == 'Absence']['Max HR']
max_hr_presence = df[df['Heart Disease'] == 'Presence']['Max HR']

# Perform independent t-test
t_statistic, p_value = stats.ttest_ind(max_hr_absence, max_hr_presence, equal_var=False) # Assuming unequal variances

print(f"Independent T-Test for Max HR between Heart Disease groups:")
print(f"T-statistic: {t_statistic:.2f}")
print(f"P-value: {p_value:.3f}")

if p_value < 0.05:
    print("Conclusion: The difference in Max HR between groups is statistically significant (p < 0.05).")
else:
    print("Conclusion: The difference in Max HR between groups is not statistically significant (p >= 0.05).")
```
