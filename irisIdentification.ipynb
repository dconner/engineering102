The Iris flower data set or Fisher's Iris data set is a multivariate data set introduced by the British statistician, eugenicist, and biologist Ronald Fisher in his 1936 paper The use of multiple measurements in taxonomic problems as an example of linear discriminant analysis.[[1](https://en.wikipedia.org/wiki/Iris_flower_data_set#:~:text=The%20Iris%20flower%20data%20set,example%20of%20linear%20discriminant%20analysis.)] It is sometimes called Anderson's Iris data set because Edgar Anderson collected the data to quantify the morphologic variation of Iris flowers of three related species.[2] Two of the three species were collected in the Gasp√© Peninsula "all from the same pasture, and picked on the same day and measured at the same time by the same person with the same apparatus".[3] Fisher's paper was published in the journal, the Annals of Eugenics, creating controversy about the continued use of the Iris dataset for teaching statistical techniques today.

The data set consists of 50 samples from each of three species of Iris (Iris setosa, Iris virginica and Iris versicolor). Four features were measured from each sample: the length and the width of the sepals and petals, in centimeters. Based on the combination of these four features, Fisher developed a linear discriminant model to distinguish the species from each other.

```python
#import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import load_iris
import numpy as np

# Load Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Create colormap based on petal width
cmap = plt.get_cmap('viridis')
norm = plt.Normalize(X[:, 3].min(), X[:, 3].max())  # Normalize petal width
colors = cmap(norm(X[:, 3]))  # Get colors for each data point

# Create plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot data points
scatter = ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=colors, s=50)

# Add colorbar for petal width
cbar = fig.colorbar(scatter, shrink=0.5, aspect=5)
cbar.set_label('Petal Width')

# Set labels
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
ax.set_zlabel('Petal Length')
ax.set_title('Iris Dataset: 3D Scatter Plot with Petal Width Shading')

# Show plot
plt.show()
```

```python
# prompt: using 3 as the K, run a k means cluster from sklearn on the dataset\

from sklearn.cluster import KMeans

# Initialize KMeans model with k=3
km = KMeans(n_clusters=3)

# Fit the model to the data
km.fit(X)

# Print cluster labels
print(km.labels_)
```

# Implementing K-Means Clustering

```python
# prompt: create a new dataset with the original 4 columns and add a 5th column corresponding to the km label

import numpy as np
labeled_iris = np.column_stack((X, km.labels_))
#print(labeled_iris)
```

```python
# prompt: replot this dataset, use the 5th column as the color, add a red data point for the kmean cluster center

import matplotlib.pyplot as plt
# Create colormap based on cluster labels
cmap = plt.get_cmap('viridis')
norm = plt.Normalize(labeled_iris[:, 4].min(), labeled_iris[:, 4].max())  # Normalize cluster labels
colors = cmap(norm(labeled_iris[:, 4]))  # Get colors for each data point

# Create plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot data points
scatter = ax.scatter(labeled_iris[:, 0], labeled_iris[:, 1], labeled_iris[:, 2], c=colors, s=50)

# Add red data points for kmeans cluster centers
ax.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], km.cluster_centers_[:, 2], c='red', s=200, marker='x')

# Add colorbar for cluster labels
cbar = fig.colorbar(scatter, shrink=0.5, aspect=5)
cbar.set_label('Cluster Label')

# Set labels
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
ax.set_zlabel('Petal Length')
ax.set_title('Iris Dataset: 3D Scatter Plot with KMeans Clusters')

# Show plot
plt.show()
```

```python
# prompt: plot 2D graphs for every combtionation of parameters, include a red x for the k mean clustering

import matplotlib.pyplot as plt

# Define the parameters
parameters = ["sepal length", "sepal width", "petal length", "petal width"]

# Create a figure with subplots
fig, axes = plt.subplots(4, 4, figsize=(16, 16))

# Loop through all combinations of parameters
for i, param1 in enumerate(parameters):
  for j, param2 in enumerate(parameters):
    # Skip if same parameter
    if i == j:
      continue

    # Plot the data points
    axes[i, j].scatter(X[:, i], X[:, j], c=colors, s=50)

    # Plot the k means cluster centers
    axes[i, j].scatter(km.cluster_centers_[:, i], km.cluster_centers_[:, j], c='red', s=200, marker='x')

    # Set labels
    axes[i, j].set_xlabel(param1)
    axes[i, j].set_ylabel(param2)

# Show the plot
plt.show()
```
